```python
# Step1. matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step2. 6個のグラフを表示する領域を作成します。Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
fig, axes = plt.subplots(2, 3, tight_layout=True)

# Step3. 各領域に、折れ線グラフ、sin関数、ヒストグラム、散布図のデータを表示するコードを記述します。

## 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

## sin関数
x0 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x0)
axes[0,1].plot(x0, y1, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))

## ヒストグラム
x0 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x0, bins=30)

## 散布図
np.random.seed(3)
x1 = 4 + np.random.normal(0, 2, 60)
y0 = 4 + np.random.normal(0, 2, len(x1))

# プロットする
axes[1,0].scatter(x1, y0)

# Step4. グラフを表示させる行「plt.show()」を書く
plt.show()
```
